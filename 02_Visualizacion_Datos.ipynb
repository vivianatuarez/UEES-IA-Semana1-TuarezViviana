```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

titanic = sns.load_dataset('titanic')

# Matplotlib
plt.figure(figsize=(10,6))
titanic['age'].hist(bins=30, alpha=0.7)
plt.title('Distribución de Edades - Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.savefig('02_Laboratorio/images/edad_titanic.png')
plt.show()

# Seaborn
sns.boxplot(x='class', y='fare', data=titanic)
plt.title('Tarifa por Clase')
plt.savefig('02_Laboratorio/images/boxplot_clase.png')
plt.show()

# Plotly (interactivo)
fig = px.scatter(titanic, x='age', y='fare', color='survived', size='pclass')
fig.show()
```
